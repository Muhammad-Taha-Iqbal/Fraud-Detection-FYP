```python
def create_schema(tx):
    queries = [
        "CREATE CONSTRAINT IF NOT EXISTS FOR (u:User) REQUIRE u.id IS UNIQUE",
        "CREATE CONSTRAINT IF NOT EXISTS FOR (t:Transaction) REQUIRE t.id IS UNIQUE",
        "CREATE CONSTRAINT IF NOT EXISTS FOR (m:Merchant) REQUIRE m.id IS UNIQUE"
    ]
    
    for query in queries:
        tx.run(query)
        print(f"Executed: {query}")  # This will print each executed query

# Run Schema Setup
with driver.session() as session:
    session.execute_write(create_schema)

print("Schema created successfully!")

```

```python
def insert_data(tx, user_id, transaction_id, merchant_id, amount, fraud_flag):
    query = """
    MERGE (u:User {id: $user_id})
    MERGE (t:Transaction {id: $transaction_id, amount: $amount, fraud_flag: $fraud_flag})
    MERGE (m:Merchant {id: $merchant_id})
    MERGE (u)-[:MADE]->(t)
    MERGE (t)-[:TO]->(m)
    """
    tx.run(query, user_id=user_id, transaction_id=transaction_id, merchant_id=merchant_id, amount=amount, fraud_flag=fraud_flag)
```

```python
import networkx as nx
from rdflib import Graph, Namespace, URIRef
import matplotlib.pyplot as plt

# Step 1: Load RDF Knowledge Graph
rdf_graph = Graph()
rdf_graph.parse("FraudDetection.rdf", format="turtle")  # Try Turtle
rdf_graph.serialize(destination="Fixed_FraudDetection.rdf", format="xml")

# Step 2: Convert RDFLib graph to NetworkX graph
nx_graph = nx.Graph()
```

```python
# Add edges to the graph (all triples in the RDF file)
for s, p, o in rdf_graph:
    nx_graph.add_edge(s, o, label=p)

# Step 3: Function to visualize relationships of a specific entity
def visualize_entity_relationships(entity_uri, depth=1, edge_limit=50):
    """
    Visualize the relationships of a specific entity in the knowledge graph.

    Args:
    - entity_uri (str): The URI of the entity to trace.
    - depth (int): How deep the relationships should go (1 = direct connections, 2 = neighbors of neighbors, etc.).
    - edge_limit (int): Limit on the number of edges to visualize for performance.
    """
    subgraph = nx.ego_graph(nx_graph, entity_uri, radius=depth)  # Get nodes/edges within 'depth' distance from the entity
    limited_edges = list(subgraph.edges(data=True))[:edge_limit]  # Limit the number of edges
    limited_graph = nx.Graph()

    # Add limited edges to the graph
    for s, o, data in limited_edges:
        limited_graph.add_edge(s, o, label=data['label'])

    # Visualize the subgraph
    plt.figure(figsize=(12, 8))  # Adjust the figure size
    pos = nx.spring_layout(limited_graph, seed=42)  # Position nodes using the spring layout
    nx.draw(limited_graph, pos, with_labels=True, node_size=500, font_size=8, font_weight='bold')
    edge_labels = nx.get_edge_attributes(limited_graph, 'label')
    nx.draw_networkx_edge_labels(limited_graph, pos, edge_labels=edge_labels, font_size=7, font_color='red')
    plt.title(f'Relationships for Entity: {entity_uri}')
    plt.show()

# Step 4: Call the function to visualize relationships for a specific entity
# Replace the entity URI with one from your graph (e.g., http://example.org/incident/Incident_001)
entity_to_trace = URIRef("http://example.org/fraud#Transaction_T054")  # Change this to the actual URI in your RDF graph
visualize_entity_relationships(entity_uri=entity_to_trace, depth=2, edge_limit=30)
```

```python
# Step 4: Define the specific entity URI
entity_to_trace = URIRef("http://example.org/fraud#Account_A032")

# Step 5: Filter only nodes connected to the specific entity
depth = 2  # Limit to nodes 2 hops away
nodes_to_trace = {entity_to_trace}

# Add nodes at each depth
for _ in range(depth):
    new_nodes = set()
    for node in nodes_to_trace:
        if node in nx_graph:
            neighbors = set(nx_graph.neighbors(node))
            new_nodes.update(neighbors)
    nodes_to_trace.update(new_nodes)

# Step 6: Subgraph containing only relevant nodes
subgraph = nx_graph.subgraph(nodes_to_trace)

# Step 7: Set up the plot
plt.figure(figsize=(50, 50))

# Step 8: Use force-directed layout
pos = nx.spring_layout(subgraph, k=0.2, iterations=50, seed=42)  # Reduce k to separate nodes

# Step 9: Draw the nodes
node_colors = ['lightcoral' if node == entity_to_trace else 'skyblue' for node in subgraph.nodes]
nx.draw_networkx_nodes(subgraph, pos, node_size=1000, node_color=node_colors, alpha=0.9)

# Step 10: Draw the edges
nx.draw_networkx_edges(subgraph, pos, width=2, alpha=0.5, edge_color="gray")

# Step 11: Draw the labels for nodes
node_labels = {node: str(node).split('#')[-1] for node in subgraph.nodes}  # Show only the last part of the URI
nx.draw_networkx_labels(subgraph, pos, labels=node_labels, font_size=10, font_color='black')

# Step 12: Draw the edge labels (only for immediate neighbors)
edge_labels = nx.get_edge_attributes(subgraph, 'label')
filtered_edge_labels = {key: str(value).split('#')[-1] for key, value in edge_labels.items()}  # Clean URI display
nx.draw_networkx_edge_labels(subgraph, pos, edge_labels=filtered_edge_labels, font_size=8, font_color='red')

# Title and axis cleanup
plt.title("Knowledge Graph for Account_A032 (Depth = 2)")
plt.axis('off')
plt.show()
```

```python
from PyQt6.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QTextEdit, QLabel
from PyQt6.QtWebEngineWidgets import QWebEngineView
from neo4j import GraphDatabase
import networkx as nx
from pyvis.network import Network

# Neo4j Connection Details
URI = "neo4j://localhost:7687"
USERNAME = "neo4j"
PASSWORD = "12345678"

class FraudDetectionGUI(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Fraud Detection System")
        self.setGeometry(100, 100, 900, 600)

        # Layout
        layout = QVBoxLayout()
        
        # Query Box
        self.query_label = QLabel("Enter Custom Query:")
        layout.addWidget(self.query_label)
        
        self.query_box = QTextEdit()
        self.query_box.setPlaceholderText("MATCH (n)-[r]->(m) RETURN n, r, m LIMIT 20;")
        layout.addWidget(self.query_box)

        # Run Query Button
        self.run_button = QPushButton("Run Query & Show Graph")
        self.run_button.clicked.connect(self.run_fraud_query)
        layout.addWidget(self.run_button)

        # Graph Visualization
        self.graph_view = QWebEngineView()
        layout.addWidget(self.graph_view)

        self.setLayout(layout)

    def run_fraud_query(self):
        """Runs the user query and visualizes the graph."""
        query = self.query_box.toPlainText()
        if not query.strip():
            query = "MATCH (n)-[r]->(m) RETURN n, r, m LIMIT 20;"
        
        # Fetch data from Neo4j
        with GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD)) as driver:
            with driver.session() as session:
                results = session.run(query)
                nodes = set()
                edges = []

                for record in results:
                    n, r, m = record["n"], record["r"], record["m"]
                    nodes.add(n["id"])
                    nodes.add(m["id"])
                    edges.append((n["id"], m["id"], r.type))

        # Build Network Graph
        G = nx.DiGraph()
        for node in nodes:
            G.add_node(node, title=node)

        for source, target, rel in edges:
            G.add_edge(source, target, label=rel)

        # Create PyVis Graph
        net = Network(height="500px", width="100%", directed=True)
        net.from_nx(G)
        net.show("fraud_graph.html")

        # Display in Web View
        self.graph_view.setUrl("fraud_graph.html")

if __name__ == "__main__":
    app = QApplication([])
    window = FraudDetectionGUI()
    window.show()
    app.exec()
```
